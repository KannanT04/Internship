In case if any product has less than 3 pages in search results then scrape all the products available under that product name. Details to be scraped are: "Brand Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
driver=webdriver.Chrome()
driver.get('https://www.amazon.in/')
```

```python
search=driver.find_element(By.XPATH,'//div[@class="nav-search-field "]//input')
search.send_keys(item)
```

```python
search_button=driver.find_element(By.XPATH,'//div[@class="nav-search-submit nav-sprite"]//span')
search_button.click()
```

```python
#scrapping product urls
product_urls=[]
start=0
end=3
for page in range(start,end):
    url=driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for i in url:
        product_urls.append(i.get_attribute('href'))
    next_button=driver.find_element(By.XPATH,'//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
    next_button.click()
    time.sleep(2)
```

```python
len(product_urls)
```

```python
Brand=[]
Name_of_the_product=[]
Price=[]
Return_Or_Exchange=[]
Expected_Delivery=[]
Availability=[]
Product_URL=[]
```

```python
for url in product_urls:
    driver.get(url)
    time.sleep(2)
    
    try:
        brand=driver.find_element(By.XPATH,'//*[@id="productOverview_feature_div"]/div/table/tbody/tr[1]/td[2]/span')
        Brand.append(brand.text)
    except NoSuchElementException:
        Brand.append('-')
    
    try:
        name=driver.find_element(By.XPATH,'//*[@id="title_feature_div"]/div/h1/span')
        Name_of_the_product.append(name.text)
    except NoSuchElementException:
        Name_of_the_product.append('-')
        
    try:
        price=driver.find_element(By.XPATH,'//*[@id="corePriceDisplay_desktop_feature_div"]/div/span[3]/span[2]/span[2]')
        Price.append(price.text)
    except NoSuchElementException:
        Price.append('-')
        
    try:
        delivery=driver.find_element(By.XPATH,'//div[@class="a-spacing-base"]/span/span[1]')
        Expected_Delivery.append(delivery.text)
    except NoSuchElementException:
        Expected_Delivery.append('-')
        
    try:
        avail=driver.find_element(By.XPATH,'//*[@id="availabilityInsideBuyBox_feature_div"]/div/div/span')
        Availability.append(avail.text)
    except NoSuchElementException:
        Availability.append('-')
```

# 5. Write a program to scrap geospatial coordinates (latitude, longitude) of a city searched on google maps.

```python
driver=webdriver.Chrome()
driver.get('https://www.google.com/maps/@10.8211663,78.2897649,7z?entry=ttu')
```

```python
sr=driver.find_element(By.XPATH,'//*[@id="searchboxinput"]')
sr.send_keys('Chennai')
```

```python
bt=driver.find_element(By.XPATH,'//*[@id="searchbox-searchbutton"]')
bt.click()
```

```python
try:
    url_string = driver.current_url
    print("URL Extracted: ", url_string)
    lat_lng = re.findall(r'@(.*)data',url_string)
    if len(lat_lng):
        lat_lng_list = lat_lng[0].split(",")
        if len(lat_lng_list)>=2:
            lat = lat_lng_list[0]
            lng = lat_lng_list[1]
        print("Latitude = {}, Longitude = {}".format(lat, lng))

except Exception as e:
        print("Error: ", str(e))
```

# 9. Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in “London” location. You have to scrape hostel name, distance from city centre, ratings, total reviews, overall reviews, privates from price, dorms from price, facilities and property description.

```python
driver=webdriver.Chrome()
```

```python
driver.get('https://www.hostelworld.com/')
```

```python
search=driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[2]/div[2]/div/div/div/div[1]/div[1]/div/div[2]/input')
search.send_keys('London, England')
```

```python
hotel_urls=[]
    
url=driver.find_elements(By.XPATH,'//a[@class="property-card-container horizontal"]')
for i in url:
    hotel_urls.append(i.get_attribute('href'))
```

```python
len(hotel_urls)
```

```python
Hotel_Name=[]
Distance_From_City_Centre=[]
Ratings=[]
Total_Reviews=[]
Overall_Reviews=[]
Private_From_price=[]
Doms_From_Price=[]
Facilities=[]
Description=[]
```

```python
for url in hotel_urls:
    driver.get(url)
    time.sleep(2)
    
    try:
        name=driver.find_element(By.XPATH,'//div[@class="title-2"]')
        Hotel_Name.append(name.text)
    except NoSuchElementException:
        Hotel_Name.append('-')
        
    try:
        rating=driver.find_element(By.XPATH,'//div[@class="score orange big"]')
        Ratings.append(rating.text)
    except NoSuchElementException:
        Ratings.append('-')
        
    try:
        tr=driver.find_element(By.XPATH,'//div[@class="reviews"]')
        Total_Reviews.append(tr.text)
    except NoSuchElementException:
        Total_Reviews.append('-')
        
    try:
        dr=driver.find_element(By.XPATH,'//div[@class="description-container"]/div/div[2]')
        Description.append(dr.text)
    except NoSuchElementException:
        Description.append('-')
        
    try:
        fl=driver.find_element(By.XPATH,'//*[@id="facilities-section"]')
        Facilities.append(fl.text)
    except NoSuchElementException:
        Facilities.append('-')
```

```python
try:
    dt=driver.find_elements(By.XPATH,'//span[@class="distance-description"]')
    for i in dt:
        Distance_From_City_Centre.append(i.text)
except NoSuchElementException:
    Distance_From_City_Centre.append('-')
```
