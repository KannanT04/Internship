# Survival Rate

```python
survival_rate=df1['Survived'].mean()*100
print('Overall Survival Rate',survival_rate)
plt.figure(figsize=(8,6))
sns.barplot(x=['Survived','Not Survived'],y=[survival_rate,100-survival_rate])
plt.title('Overall Survival Rate')
plt.ylabel('Percentage')
plt.show()
```

# Feature Selection

```python
features=['Pclass','Sex','Age','Fare','Embarked','Family Size']
target='Survived'
```

```python
x=df1[features]
y=df1[target]
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif=pd.DataFrame()
vif['Features']=x.columns
vif['VIF values']=[variance_inflation_factor(x.values,i) for i in range(len(x.columns))]

vif
```

```python
scaler=StandardScaler()
X=scaler.fit_transform(x)
X
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.model_selection import cross_val_score,GridSearchCV
```

```python
lr=LogisticRegression()
maxAcc=0
rs=0
for i in range(1000):
    x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=i)
    lr.fit(x_train,y_train)
    pred=lr.predict(x_test)
    ac=accuracy_score(y_test,pred)
    if ac>maxAcc:
        maxAcc=ac
        rs=i
print('Best Score:',maxAcc)
print('Random State:',rs)
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=150)
```

# Logistic Regression

```python
lg=LogisticRegression()
lg.fit(x_train,y_train)
pred=lg.predict(x_test)
acc_score=accuracy_score(y_test,pred)
class_report=classification_report(y_test,pred)
cv_score=cross_val_score(lg,X,y,cv=5).mean()
result=acc_score-cv_score
print('Accuracy Score:',acc_score)
print('Confusion Matrix:',confusion_matrix(y_test,pred))
print('Classification Matrix:',class_report)
print('Cross Validation:',cv_score)
print("Result:",result)
```

# Decision Tree Classifier

```python
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
predict=dt.predict(x_test)
acc=accuracy_score(y_test,predict)
class_report=classification_report(y_test,pred)
cv_score=cross_val_score(dt,X,y,cv=5).mean()
Result=acc-cv_score
print('Accuracy_Score',acc)
print('Confusion Matrix',confusion_matrix(y_test,predict))
print('Classification Report',class_report)
print('Cross Validation:',cv_score)
print("Result:",Result)
```

# Random Forest Classifier

```python
rc=RandomForestClassifier()
rc.fit(x_train,y_train)
pred=rc.predict(x_test)
acc_score=accuracy_score(y_test, pred)
class_report=classification_report(y_test,pred)
cv_score=cross_val_score(rc,X,y,cv=5).mean()
result=acc_score-cv_score
print('Accuracy_Score',acc_score)
print('Confusion Matrix',confusion_matrix(y_test,pred))
print('Classification Report',class_report)
print('Cross Validation:',cv_score)
print("Result:",result)
```

# Gradient Boosting Classifier

```python
gb=GradientBoostingClassifier()
gb.fit(x_train,y_train)
pred=gb.predict(x_test)
acc_score=accuracy_score(y_test,pred)
class_report=classification_report(y_test,pred)
cv_score=cross_val_score(lg,X,y,cv=5).mean()
result=acc_score-cv_score
print('Accuracy Score:',acc_score)
print('Confusion Matrix:',confusion_matrix(y_test,pred))
print('Classification Matrix:',class_report)
print('Cross Validation:',cv_score)
print("Result:",result)
```
